```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
import numpy as np

## Distribution Plots
def plot_distribution(column_data):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.hist(column_data, bins=30)
    plt.title("Histogram")

    plt.subplot(1, 2, 2)
    stats.probplot(column_data, dist="norm", plot=plt)
    plt.title("Q-Q Plot")

    plt.show()

## Check if the data is normally distrubuted
def check_normality(column_data, alpha=0.05):
    _, p_value = stats.shapiro(column_data)

    if p_value > alpha:
        return "The data appears to be normally distributed."
    else:
        return "The data does not appear to be normally distributed."

## Check the normality of the data
def analyse_normality(dataframe, column_name):
    column_data = dataframe[column_name]

    ## Visual inspection
    plot_distribution(column_data)

    ## Statistical normality test prior to removing the outliers
    result = check_normality(column_data)
    
    return result


## Get insight around the distribution of the data
def get_insight(dataframe, column_name):
    df = pd.DataFrame(dataframe[column_name])
    n_obs = len(df)
    min_obs = df.min()
    mean_obs = df.mean()
    max_obs = df.max()
    
    ## Calculate the Z-Score for each data point
    df["Z-Score"] = np.abs((df - df.mean()) / df.std())

    # Create a list of Z-scores to analyze
    z_scores = [1, 2, 3]
    
    results = []
    
    for z_score in z_scores:
        within_z_score = df[df["Z-Score"] <= z_score]
        z_score_maxval = within_z_score.max().iloc[0]
        z_score_minval = within_z_score.min().iloc[0]
        z_score_len = len(within_z_score)
        z_score_percent = (z_score_len / n_obs) * 100
        
        result = f"""\n********** Within {z_score} Z-Score
Number of observations: {z_score_len}
Percent of observations: {z_score_percent}
Min value: {z_score_minval}
Max value: {z_score_maxval}
"""
        results.append(result)
    
    # Prepare the formatted output
    formatted_output = f"""
********** Original
Total number of observations: {n_obs}
Min value: {min_obs[0]}
Mean value: {mean_obs[0]}
Max value: {max_obs[0]}
{''.join(results)}
"""
    return formatted_output


## Remove the outliers
def remove_outliers(dataframe, column_name, z_score_threshold):
    
    dataframe["Z-Score"] = np.abs((dataframe[column_name] - dataframe[column_name].mean()) / dataframe[column_name].std())
    
    ## Calculate the lower and upper bounds of the data within 1 Z-Score above the user input Z-Score
    z_score_plus_min = dataframe[dataframe["Z-Score"] <= (z_score_threshold + 1)][column_name].min()
    z_score_plus_max = dataframe[dataframe["Z-Score"] <= (z_score_threshold + 1)][column_name].max()
    cap_vals = (z_score_plus_min, z_score_plus_max)
    
    # Detect outliers
    outliers = dataframe[dataframe["Z-Score"] > z_score_threshold]

    # Remove outliers
    clean = dataframe[dataframe["Z-Score"] <= z_score_threshold]
    clean = clean.drop(columns="Z-Score")
    return outliers, clean, cap_vals
```
